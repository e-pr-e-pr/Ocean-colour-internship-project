```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

for i, header in enumerate(headers_to_plot): 
    col = i // 1
    row = i % 1
    
    x_values = ds_chl_aug['insitu(chl)']
    y_values = ds_chl_aug[header]
    

  #  min_val = min(min(x_values), min(y_values)) # can calc this but i think we always want it from 0 no?
    max_val = max(max(x_values), max(y_values)) + 0.6 # ad a lil bit to make sure is in range, other wise cuts it off. have to kinda play it by ear and see what the plot looks like to settle on a number

    axs[col].scatter(ds_chl_aug['insitu(chl)'], ds_chl_aug[header], marker = 'o', label = 'data')

    # Set equal aspect ratio with buffer
    axs[col].set_xlim(0, max_val)
    axs[col].set_ylim(0, max_val)

    axs[col].axline([0,0], [1, 1], color='red', linestyle='-', label='1:1 line')


    axs[col].set_title(f'{header} vs in-situ chl-a, 25.08.2016')
    axs[col].tick_params(axis='x', rotation=45)
    axs[col].set_xlabel('In-situ Chlorophyll data [mg/L]') 
    axs[col].set_ylabel(f'{header} Chlorophyll [mg/L]')
    axs[col].legend(loc='lower right')

plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.5, hspace=0.5)
plt.tight_layout()
#plt.savefig('C:/Users/a86963/Documents/results/algorithm_vs_insitu_plots.jpeg')
plt.show()
```
