```python
fig, axs = plt.subplots(4, 2, figsize=(10, 20))

for i, header in enumerate(columns_to_plot): 
    col = i // 2 
    row = i % 2
    
    mask = ~np.isnan(ds['Chla(Spect)']) & ~np.isnan(ds[header]) # this part checks for the values that will be plotted i.e. where there eaxists data for both the in situ and alg data. this is necessary to set the proper aspect ratio, otherwise it does it over the whole dataset, which might be wrong sce not all the alg results have in situ data to comapre to 
    x_values = ds['Chla(Spect)'][mask]
    y_values = ds[header][mask]
    

  #  min_val = min(min(x_values), min(y_values)) # can calc this but i think we always want it from 0 no?
    max_val = max(max(x_values), max(y_values)) + 1 # plus one to add an extra tick to make suer the points are in range

    axs[col, row].scatter(ds['Chla(Spect)'], ds[header], marker = 'o', label = 'data')

    # Set equal aspect ratio with buffer
    axs[col, row].set_xlim(0, max_val)
    axs[col, row].set_ylim(0, max_val)

    axs[col, row].axline([0,0], [1, 1], color='red', linestyle='-', label='1:1 line')


    axs[col, row].set_title(f'{header} vs in-situ chl-a')
    axs[col, row].tick_params(axis='x', rotation=45)
    axs[col, row].set_xlabel('In-situ Chlorophyll data [mg/L]') 
    axs[col, row].set_ylabel(f'{header} Chlorophyll [mg/L]')
    axs[col, row].legend(loc='lower right')

plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.5, hspace=0.5)
plt.tight_layout()
plt.savefig('C:/Users/a86963/Documents/results/algorithm_vs_insitu_plots.jpeg')
plt.show()

###

fig, axs = plt.subplots(4, 2, figsize=(10, 20))

for i, header in enumerate(columns_to_plot): 
    col = i // 2 
    row = i % 2
    
    # Filter out masked or empty values
    mask = ~np.isnan(ds['Chla(Spect)']) & ~np.isnan(ds[header])
    x_values = ds['Chla(Spect)'][mask]
    y_values = ds[header][mask]
    
    if len(x_values) == 0 or len(y_values) == 0:
        # skip if there are no valid data points
        continue

    # Calculate the overall range of x and y values for this subplot
    min_val = min(min(x_values), min(y_values))
    max_val = max(max(x_values), max(y_values)) + 1 # plus one to add an extra tick to make suer the points are in range

    axs[col, row].scatter(ds['Chla(Spect)'], ds[header], marker = 'o', label = 'data')

    # Set equal aspect ratio with buffer
    axs[col, row].set_xlim(min_val, max_val)
    axs[col, row].set_ylim(min_val, max_val)

    axs[col, row].axline([0,0], [1, 1], color='red', linestyle='-', label='1:1 line')


    axs[col, row].set_title(f'{header} vs in-situ chl-a')
    axs[col, row].tick_params(axis='x', rotation=45)
    axs[col, row].set_xlabel('In-situ Chlorophyll data [mg/L]') 
    axs[col, row].set_ylabel(f'{header} Chlorophyll [mg/L]')
    axs[col, row].legend()

plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.5, hspace=0.5)
plt.tight_layout()
plt.show()
```
